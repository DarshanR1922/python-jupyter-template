### Exercise 5: Subplot Layout

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plots
x = np.linspace(0, 10, 100)
y = np.sin(x)

bar_categories = ['A', 'B', 'C', 'D']
bar_values = [5, 7, 3, 9]

scatter_x = [1, 2, 3, 4, 5]
scatter_y = [2, 4, 6, 8, 10]

pie_labels = ['Python', 'Java', 'C++', 'JavaScript']
pie_sizes = [40, 25, 20, 15]
pie_colors = ['blue', 'red', 'green', 'orange']

# Create subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot (First Subplot)
axs[0, 0].plot(x, y, color='blue')
axs[0, 0].set_title("Line Chart")

# Bar Plot (Second Subplot)
axs[0, 1].bar(bar_categories, bar_values, color=['red', 'blue', 'green', 'purple'])
axs[0, 1].set_title("Bar Chart")

# Scatter Plot (Third Subplot)
axs[1, 0].scatter(scatter_x, scatter_y, color='black', marker='o')
axs[1, 0].set_title("Scatter Plot")

# Pie Chart (Fourth Subplot)
axs[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%', colors=pie_colors, startangle=140, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title("Pie Chart")

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```
